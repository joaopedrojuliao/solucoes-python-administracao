1. Classificação de Inadimplência: Peça o número de dias de atraso de uma
fatura. Se for 0, "Pagamento em dia". Até 30 dias, "Juros de 2%". Acima de 30
dias, "Encaminhar para cobrança jurídica".

```python
dias_atraso = int(input("Digite o número de dias de atraso: "))
if(dias_atraso == 0):
    print("Pagamento em dia")
elif(dias_atraso <= 30):
    print("Juros de 2%")
else:
    print("Encaminhar para cobrança jurídica")

```

2. Bônus de Performance: Peça a meta batida por um vendedor (em %). Se <
80%, "Sem bônus". Entre 80% e 100%, "Bônus de R$ 500". Acima de 100%,
"Bônus de R$ 1.500".

```python
meta_batida = int(input("Digite a meta batida: "))
if(meta_batida < 80):
    print("Sem bônus")
elif(meta_batida >= 80 and meta_batida <= 100):
    print("Bônus de R$ 500")
else:
    print("Bônus de R$ 1.500")
```

3. Prioridade de Atendimento: Peça a idade do cliente e se ele possui
deficiência (Sim/Não). Se idade >= 60 ou deficiência == "Sim", imprima "Fila
Prioritária", caso contrário, "Fila Comum".

```python
idade = int(input("Digite a idade do cliente: "))
deficiencia = input("O cliente possui deficiência? (Sim/Não): ")

if idade >= 60 or deficiencia.lower() == "sim":
    print("Fila Prioritária")
else:
    print("Fila Comum")
```

4. Viabilidade de Investimento: Peça o ROI (Retorno sobre Investimento)
esperado. Se ROI > 20%, "Investimento Recomendado". Entre 10% e 20%,
"Risco Moderado". Abaixo de 10%, "Investimento Inviável".

```python
roi = float(input("Digite o ROI esperado (%): "))

if roi > 20:
    print("Investimento Recomendado")
elif roi >= 10 and roi <= 20:
    print("Risco Moderado")
else:
    print("Investimento Inviável")
```

5. Soma de Horas Extras: Você tem uma lista de horas extras de 10
funcionários: horas = [2, 5, 0, 8, 3, 4, 1, 0, 10, 2]. Use um for para calcular o total de
horas que a empresa pagará no mês.


```python
horas = [2, 5, 0, 8, 3, 4, 1, 0, 10, 2]
total_horas = 0

for hora in horas:
    total_horas += hora

print(f"Total de horas extras a pagar: {total_horas}")
```

6. Filtro de Estoque Crítico: Dada a lista estoque = [50, 5, 12, 3, 45, 8], use um
loop para imprimir apenas os valores menores que 10, alertando: "Item abaixo
do estoque mínimo!".

```python
estoque = [50, 5, 12, 3, 45, 8]

for x in estoque:
    if x < 10:
        print(f"Item {x} abaixo do estoque mínimo!")
```

7. Reajuste Salarial: Dada a lista salarios = [2000, 3500, 4000, 1800], crie uma nova
lista onde cada salário é multiplicado por 1.05 (aumento de 5%) usando um
loop.

```python
salarios = [2000, 3500, 4000, 1800]
novos_salarios = []

for salario in salarios:
    novos_salarios.append(salario * 1.05)

print(f"Salários originais: {salarios}")
print(f"Novos salários (com aumento de 5%): {novos_salarios}")
```

8. Contador de Faltas: Dada a lista de presenças presenca = ["P", "P", "F", "P", "F",
"P", "P"], conte quantas faltas ("F") ocorreram na semana.

```python
presenca = ["P", "P", "F", "P", "F", "P", "P"]
contador_faltas = 0

for status in presenca:
    if status == "F":
        contador_faltas += 1

print(f"Número de faltas na semana: {contador_faltas}")
```
